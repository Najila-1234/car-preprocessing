```python
import pandas as pd
df=pd.read_csv('train-data.csv')
df
```

```python
# create dummy
dummy=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']], drop_first=True)
dummy
```

```python
dfe = pd.concat([df,dummy], axis=1)
dfe
```

```python
dfe=dfe.drop(['Location','Fuel_Type','Transmission','Owner_Type','New_Price','Name'],axis=1)
dfe
```

```python
df.dtypes
```

```python
dfe.dtypes
```

```python
# remove unit in mileage,engine,power replace with non space ,convert strng to int
dfe['Mileage']=dfe['Mileage'].str.replace('kmpl','')
dfe['Mileage']=dfe['Mileage'].str.replace('km/kg','')
dfe['Engine']=dfe['Engine'].str.replace('CC','')
dfe['Power']=dfe['Power'].str.replace('bhp','')
# replace null with 0

dfe['Power']=dfe['Power'].str.replace('null','0')
dfe
```

```python
#drop fuel type electric in x colms before training after creating dummies
dfe=dfe.drop('Fuel_Type_Electric',axis=1)
dfe
```

```python
dfe.dtypes
```

```python
# convert mileage,engine,power to float, still it exist as object using astype fn

dfe['Engine']=dfe['Engine'].astype(float)
dfe['Mileage']=dfe['Mileage'].astype(float)
dfe['Power']=dfe['Power'].astype(float)
dfe
```

```python
dfe.dtypes
```

```python
# to use fillna fn convert 0 to NaN
import numpy as np
dfe.loc[dfe.Engine==0,'Engine']=np.NaN
dfe.loc[dfe.Power==0,'Power']=np.NaN
dfe
```

```python
# fill mileage,eng,power with mean and seat with mode
dfe['Engine']=dfe['Engine'].fillna(dfe['Engine'].mean())
dfe['Mileage']=dfe['Mileage'].fillna(dfe['Mileage'].mean())
dfe['Power']=dfe['Power'].fillna(dfe['Power'].mean())
dfe['Seats']=dfe['Seats'].fillna(dfe['Seats'].mode()[0])
dfe
```

```python
dfe.isna().sum()
```

```python
# separate x and y
x=dfe.drop(['Price'],axis=1)
y=dfe['Price']
x
y
```

```python
# import linear model, use multiple linear regression
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
lin.fit(x,y)
```
